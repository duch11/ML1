```python
# Load weather data (take a look at that CSV file on GitHub or via Excel and note column names)
weather = np.loadtxt("OshawaWeather2016.csv", skiprows=1, delimiter=',')
print(weather)
```

#### Crate separate 1D arrays for each column (e.g: day, maxt, mint and so on)

```python
# your code here
day = weather[:,0]
maxTempC = weather[:,1]
minTempC = weather[:,2]
meanTempC = weather[:,3]
precipTotMM = weather[:,4]
# Test to see the correct input
index = 3
print(day[index])
print(maxTempC[index])
print(minTempC[index])
print(meanTempC[index])
print(precipTotMM[index])
# Is all OK! :)
```

#### Plot the minimum temperature as a function of day number

```python
plt.plot(day,minTempC)
```

###### IT WORKS !! :D

#### Plot the max temperature in degrees Fahrenheit

```python
maxTempF = maxTempC * 1.8 + 32
maxTempF
plt.plot(day, maxTempF)
```

#### Plot the difference between the maximum and the minimum day temperature as a function of day number

```python
diffTempMaxMinC = maxTempC - minTempC

plt.plot(day, diffTempMaxMinC)
```

#### Plot both, the minimum and maximum temperature on the same figure

```python
maxTemperatureInC = plt.plot(day, maxTempC, label='Max Temperature')
minTemperatureInC = plt.plot(day, minTempC, label='Min Temperature')
plt.xlabel('day of the year')
plt.ylabel('temp Celcius')
plt.title('Min and max temp C Oshawa 2016')
plt.legend()
```
